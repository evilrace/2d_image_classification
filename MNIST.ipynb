```python
# import modules
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras import layers
```

```python
num_classes = 10
input_shape = (28, 28, 1)

(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# scale image to the [0, 1] range
x_train = x_train.astype('float32') / 255
x_test  = x_test.astype('float32') / 255

# convert image shape from (28, 28) to (28, 28, 1)
x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)

# check image shape
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(y_train.shape[0], 'test samples')

y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
# convert to tf dataset from numpy dataset
train_ds = tf.data.Dataset.from_tensor_slices((x_train, y_train))
test_ds = tf.data.Dataset.from_tensor_slices((x_test, y_test))
train_ds = train_ds.batch(128).prefetch(128)
test_ds = test_ds.batch(128).prefetch(128)
```

```python
model = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        layers.Conv2D(32, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(64, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Flatten(),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation="softmax"),
    ]
)
model.summary()
```

```python
batch_size = 32
epochs = 15

optimizer = keras.optimizers.Adam(1e-3)
loss = keras.losses.CategoricalCrossentropy()
metrics = [keras.metrics.CategoricalAccuracy()]


model.compile(loss=loss, optimizer=optimizer, metrics=metrics)

model.fit(x_train, y_train, epochs=epochs)
```

```python
score = model.evaluate(x_test, y_test, verbose=0)
print("Test loss:", score[0])
print("Test accuracy:", score[1])
```
